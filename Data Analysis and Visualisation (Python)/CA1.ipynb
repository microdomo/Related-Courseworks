```python
#Histogram
new_acute_facils = []
new_ch_facils = []
new_ph_facils = []
new_ih_facils = []
new_nh_facils = []

#Data for Facilities
acute_facils = data[data['facility_type_a'] == 'Acute']['no_of_facilities']
ch_facils = data[data['facility_type_a'] == 'Community Hospitals']['no_of_facilities']
ph_facils = data[data['facility_type_a'] == 'Psychiatric Hospitals']['no_of_facilities']
ih_facils = data[data['facility_type_a'] == 'Inpatient Hospices']['no_of_facilities']
nh_facils = data[data['facility_type_a'] == 'Nursing Homes']['no_of_facilities']

for a in range(len(acute_facils)-1):
    new_acute = acute_facils[a+1] - acute_facils[a]
    new_acute_facils.append(new_acute)

for c in range(len(ch_facils)-1):
    new_ch = ch_facils[c+1] - ch_facils[c]
    new_ch_facils.append(new_ch)

for p in range(len(ph_facils)-1):
    new_ph = ph_facils[p+1] - ph_facils[p]
    new_ph_facils.append(new_ph)
    
for i in range(len(ih_facils)-1):
    new_ih = ih_facils[i+1] - ih_facils[i]
    new_ih_facils.append(new_ih)
    
for n in range(len(nh_facils)-1):
    new_nh = nh_facils[n+1] - nh_facils[n]
    new_nh_facils.append(new_nh)

alladded = new_acute_facils + new_ch_facils + new_ph_facils + new_ih_facils + new_nh_facils

#Plotting histogram
plt.figure(figsize=(7,7))

plt.hist(alladded,facecolor='mediumpurple')
plt.title('Frequency of changes in number of facilities every year',size=13)
plt.xlabel('Changes in number of facilities',size=13)
plt.ylabel('Number of years',size=13)

plt.show()
```

- The least frequent changes in number of facilities is 3 facilities a year at 0.
- An increase of 4, 5 and 6 facilities all have the same frequency of 1.
- A decrease of 2 facilities also have the same frequency of 1.

### 2. Residential Long-Term Care Admissions
This dataset tells us the types of inpatient facilities and the admission count for each of it over the years.

```python
#Grouped bar chart
import numpy as np
import matplotlib.pyplot as plt

#Read the csv file using genfromtxt
filename = "residential-long-term-care-admissions.csv"
data = np.genfromtxt(filename, dtype=['int', 'U18', 'int'], 
                     delimiter=',', names=True)

#Print out total rows of data in the file
print(f'There are altogether {len(data)} rows in this dataset.\n')

#Print out the number of years of data captured
data_years = data['year']
years = np.unique(data_years)

print(f'There are {len(years)} years of data captured from {years[0]} to {years[-1]}.\n')

#Print out the types of facilities
LT_type = data['type']
LTF_type = np.unique(LT_type)

print(f'There are {len(LTF_type)} types of facilities: {LTF_type[0]} and {LTF_type[-1]}\n')

#Data
h_admissioncount = data[data['type'] == 'Inpatient Hospices']['count']
n_admissioncount = data[data['type'] == 'Nursing Homes']['count']
n_years = data[data['type'] == 'Nursing Homes']['year']


#Print out the highest and lowest count for inpatient hospices
print("=============================================")
print("             Inpatient Hospices                ")
print("=============================================  ")
h_highest_count = h_admissioncount.max()
hargmax = h_admissioncount.argmax()
h_highest_year = data[data['type'] == 'Inpatient Hospices'][hargmax][0]

print(f'Year with the highest admission count: {h_highest_year}')
print(f'Admission Count: {h_highest_count} ')

h_lowest_count = h_admissioncount.min()
hargmin = h_admissioncount.argmin()
h_lowest_year = data[data['type'] == 'Inpatient Hospices'][hargmin][0]

print(f'\nYear with the lowest admission count: {h_lowest_year}')
print(f'Admission Count: {h_lowest_count} ')

hdifference = h_admissioncount[-1]-h_admissioncount[0]

if hdifference > 0:
    hincredecre = 'increase'
else:
    hincredecre = 'decrease'

print(f'\nOverall, there is an {hincredecre} of {hdifference} admissions from {years[0]} to {years[-1]}')

#Print out the highest and lowest count for nursing homes
print("\n=============================================")
print("                Nursing Homes                  ")
print("=============================================  ")
n_highest_count = n_admissioncount.max()
nargmax = n_admissioncount.argmax()
n_highest_year = data[data['type'] == 'Nursing Homes'][nargmax][0]

print(f'Year with the highest admission count: {n_highest_year}')
print(f'Admission Count: {n_highest_count} ')

n_lowest_count = n_admissioncount.min()
nargmin = n_admissioncount.argmin()
n_lowest_year = data[data['type'] == 'Nursing Homes'][nargmin][0]

print(f'\nYear with the lowest admission count: {n_lowest_year}')
print(f'Admission Count: {n_lowest_count} ')

ndifference = n_admissioncount[-1]-n_admissioncount[0]

if ndifference > 0:
    nincredecre = 'increase'
else:
    nincredecre = 'decrease'
    
print(f'\nOverall, there is an {nincredecre} of {ndifference} admissions from {n_years[0]} to {n_years[-1]}')


#Plotting of bar chart
width = 0.4  # the width of the bars
fig, ax = plt.subplots(figsize=(12,6))

# Inpatient Hospices bar
ax.bar(years - width/2, h_admissioncount, width, label='Inpatient Hospices',color='mediumturquoise')
# Nursing Home bar
ax.bar(n_years + width/2, n_admissioncount, width, label='Nursing Homes',color='salmon')

ax.set_ylabel('Admission Count',size=13)
ax.set_xlabel('Years',size=13)
ax.set_title('Admission Count for Residential Long-Term Care from 2007 to 2019',size=13)
ax.legend()

plt.show()

print("**No bar for Nursing Homes in 2007 as there was no data given")
```

### 3. Hospital Admission Rate by Age and Sex
This dataset tells us the types of hospitals and the admission rate for each of it over the years. The admission rate is per 1000 resident population and is also grouped by the sex and age of the patients.

```python
import numpy as np
import matplotlib.pyplot as plt

#Read the csv file using genfromtxt
filename = "hospital-admission-rate-by-age-and-sex.csv"
data = np.genfromtxt(filename, dtype=['int', 'U21', 'U6', 'U15', 'float'], 
                     delimiter=',', names=True)

#Print out total rows of data in the file
print(f'There are altogether {len(data)} rows in this dataset.\n')

#Print out the number of years of data captured
data_years = data['year']
years = np.unique(data_years)

print(f'There are {len(years)} years of data captured from {years[0]} to {years[-1]}.\n')

#Print out the types of facilities
H_type = data['facility_type_a']
Hospital_type = np.unique(H_type)

print(f'There are {len(Hospital_type)} types of facility: {Hospital_type[0]}, {Hospital_type[1]} and {Hospital_type[-1]}')

#Data for Acute
print("\n=============================================")
print("                    Acute                      ")
print("=============================================  ")

acute_rate = data[data['facility_type_a'] == 'Acute']['rate']
acute_highest_rate = acute_rate.max()
acute_argmax = acute_rate.argmax()
acute_highest_rate_year = data[data['facility_type_a'] == 'Acute'][acute_argmax][0]
acute_highest_rate_gender = data[data['facility_type_a'] == 'Acute'][acute_argmax][2]
acute_highest_rate_age = data[data['facility_type_a'] == 'Acute'][acute_argmax][3]

acute_lowest_rate = acute_rate.min()
acute_argmin = acute_rate.argmin()
acute_lowest_rate_year = data[data['facility_type_a'] == 'Acute'][acute_argmin][0]
acute_lowest_rate_gender = data[data['facility_type_a'] == 'Acute'][acute_argmin][2]
acute_lowest_rate_age = data[data['facility_type_a'] == 'Acute'][acute_argmin][3]

#Print out the highest and lowest admission rate for Acute including gender and age group
print(f'Year with the highest admission rate: {acute_highest_rate_year}')
print(f'Sex and Age group of the highest admission rate: {acute_highest_rate_gender}s who are {acute_highest_rate_age} ')
print(f'Admission Rate: {acute_highest_rate} ')

print(f'\nYear with the lowest admission rate: {acute_lowest_rate_year}')
print(f'Sex and Age group of the lowest admission rate: {acute_lowest_rate_gender}s who are {acute_lowest_rate_age} ')
print(f'Admission Rate: {acute_lowest_rate} ')
    
#Data for Community Hospitals
print("\n=============================================")
print("             Community Hospitals               ")
print("=============================================  ")

ch_rate = data[data['facility_type_a'] == 'Community Hospitals']['rate']
ch_highest_rate = ch_rate.max()
ch_argmax = ch_rate.argmax()
ch_highest_rate_year = data[data['facility_type_a'] == 'Community Hospitals'][ch_argmax][0]
ch_highest_rate_gender = data[data['facility_type_a'] == 'Community Hospitals'][ch_argmax][2]
ch_highest_rate_age = data[data['facility_type_a'] == 'Community Hospitals'][ch_argmax][3]

ch_rate = data[data['facility_type_a'] == 'Community Hospitals']['rate']
ch_lowest_rate = ch_rate.min()
ch_argmin = ch_rate.argmin()
ch_lowest_rate_year = data[data['facility_type_a'] == 'Community Hospitals'][ch_argmin][0]
ch_lowest_rate_gender = data[data['facility_type_a'] == 'Community Hospitals'][ch_argmin][2]
ch_lowest_rate_age = data[data['facility_type_a'] == 'Community Hospitals'][ch_argmin][3]

#Print out the highest and lowest admission rate for Community Hospitals including gender and age group
print(f'Year with the highest admission rate: {ch_highest_rate_year}')
print(f'Sex and Age group of the highest admission rate: {ch_highest_rate_gender}s who are {ch_highest_rate_age} ')
print(f'Admission Rate: {ch_highest_rate} ')

print(f'\nYear with the lowest admission rate: {ch_lowest_rate_year}')
print(f'Sex and Age group of the lowest admission rate: {ch_lowest_rate_gender}s who are {ch_lowest_rate_age} ')
print(f'Admission Rate: {ch_lowest_rate} ')

#Data for Psychiatric Hospitals
print("\n=============================================")
print("             Psychiatric Hospitals             ")
print("=============================================  ")

ph_rate = data[data['facility_type_a'] == 'Psychiatric Hospitals']['rate']
ph_highest_rate = ph_rate.max()
ph_argmax = ph_rate.argmax()
ph_highest_rate_year = data[data['facility_type_a'] == 'Psychiatric Hospitals'][ph_argmax][0]
ph_highest_rate_gender = data[data['facility_type_a'] == 'Psychiatric Hospitals'][ph_argmax][2]
ph_highest_rate_age = data[data['facility_type_a'] == 'Psychiatric Hospitals'][ph_argmax][3]

ph_rate = data[data['facility_type_a'] == 'Psychiatric Hospitals']['rate']
ph_lowest_rate = ph_rate.min()
ph_argmin = ph_rate.argmin()
ph_lowest_rate_year = data[data['facility_type_a'] == 'Psychiatric Hospitals'][ph_argmin][0]
ph_lowest_rate_gender = data[data['facility_type_a'] == 'Psychiatric Hospitals'][ph_argmin][2]
ph_lowest_rate_age = data[data['facility_type_a'] == 'Psychiatric Hospitals'][ph_argmin][3]

#Print out the highest and lowest admission rate for Psychiatric Hospitals including gender and age group
print(f'Year with the highest admission rate: {ph_highest_rate_year}')
print(f'Sex and Age group of the highest admission rate: {ph_highest_rate_gender}s who are {ph_highest_rate_age} ')
print(f'Admission Rate: {ph_highest_rate} ')

print(f'\nYear with the lowest admission rate: {ph_lowest_rate_year}')
print(f'Sex and Age group of the lowest admission rate: {ph_lowest_rate_gender}s who are {ph_lowest_rate_age} ')
print(f'Admission Rate: {ph_lowest_rate} ')
```

```python
#Data for Acute boxplot
#set up the y values
young = data[data['age'] == '0-14 Years']
acuteyoung = young[young['facility_type_a'] == 'Acute']['rate']

middle = data[data['age'] == '15-64 years']
acutemiddle = middle[middle['facility_type_a'] == 'Acute']['rate']

oldest = data[data['age'] == '65 years & over']
acuteoldest = oldest[oldest['facility_type_a'] == 'Acute']['rate']

acute_y_values = np.array([acuteyoung, acutemiddle, acuteoldest])

#set up the labels on the x-axis
acute_x_labels = ['0-14 Years', '15-64 years', '65 years & over']

#label the Acute boxplot
plt.figure(figsize=(6,8))
plt.title('Admission Rates for the different age groups for Acute from 2009 to 2020',y=1.03)
plt.xlabel('Age Groups', size = 12)
plt.ylabel('Admission Rates', size = 12)
plt.ylim([0,400])

plt.boxplot(acute_y_values.transpose(),labels=acute_x_labels,patch_artist=True)
plt.show()
```

### Data Analysis of the Acute box plot above

The box plot shows the admission rates for each of the age groups: 0-14 years old, 15-64 years old and 65 years and above.
